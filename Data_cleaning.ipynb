# Data Cleaning

```python
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('/Users/markdetiberiis/Desktop/Patient_no_show/data.csv')
```

```python
df
```

```python
df["ScheduledDay"]= pd.to_datetime(df["ScheduledDay"]) 
```

```python
df['schedule_date'] = df['ScheduledDay'].dt.date

df['schedule_time'] = df['ScheduledDay'].dt.time
```

```python
df.drop(['ScheduledDay'],axis =1, inplace = True)
```

```python
df["AppointmentDay"]= pd.to_datetime(df["AppointmentDay"]) 
```

```python
df['appointment_date'] = df['AppointmentDay'].dt.date

df['appointment_time'] = df['AppointmentDay'].dt.time
```

```python
#get Day of the week for each appointment 
df["appointment_date"]= pd.to_datetime(df["appointment_date"]) 
df['appointment_weekday'] = df['appointment_date'].dt.day_name()
```

```python
df = df.rename(columns={'Hipertension': 'Hypertension', 'Handcap': 'Handicap', 'SMS_received': 'SMSReceived', 'No-show': 'NoShow'})
```

```python
#number of unique neighborhoods

df.Neighbourhood.unique().size
```

```python
#get Day of the week for each appointment 
df["schedule_date"]= pd.to_datetime(df["schedule_date"]) 
```

```python
#get DOW

df['schedule_weekday'] = df['schedule_date'].dt.day_name()
```

```python
df.groupby(['NoShow']).count()
```

```python
# Get the Waiting Time in Days of the Patients.
df['waiting_days'] = df['appointment_date'] - df['schedule_date']
df['waiting_days'] = df['Waiting_Time_days'].dt.days
```

```python
df['schedule_date_Y'] = df['schedule_date'].dt.year
df['schedule_date_M'] = df['schedule_date'].dt.month
df['schedule_date_D'] = df['schedule_date'].dt.day
df.drop(['schedule_date'], axis=1, inplace=True)

df['appointment_date_Y'] = df['AppointmentDay'].dt.year
df['appointment_date_M'] = df['AppointmentDay'].dt.month
df['appointment_date_D'] = df['AppointmentDay'].dt.day
df.drop(['AppointmentDay'], axis=1, inplace=True)
```
